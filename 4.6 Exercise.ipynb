## Exercise 4.6 - Combining & Exporting Data

### The script contains following points:

#### 1.  Create dictionaries of data

#### 2.  Convert dictionaries into dataframes

#### 3.  Concatenate the dataframes

```python
# Import Libraries

import pandas as pd
import numpy as np
import os
```

#### 1.  Create dictionaries of data

```python
# Define a dictionary containing January 2020 data
data1 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'purchased_meat':[0, 13, 3, 4],
    'purchased_alcohol':[1, 2, 10, 0],
    'purchased_snacks':[10, 5, 1, 7]}


# Define a dictionary containing February 2020 data
data2 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
    'purchased_meat':[0, 10, 5, 3],
    'purchased_alcohol':[2, 4, 14, 0],
    'purchased_snacks':[15, 3, 2, 6]}
```

#### 2.  Convert dictionaries of dataframes

```python
#Convert the dictionary into dataframe
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
df_1 = pd.DataFrame(data2,index=[0, 1, 2, 3])
```

```python
df
```

```python
df_1
```

#### 3. Concatenate the dataframes

```python
# Create a list that contains our dataframes
frames=[df,df_1]
```

```python
# Check the output
frames
```

```python
# Check the data types to be sure it is a list
type(frames)
```

```python
# Concatenate the dataframes with rows or columns of the same length using default options 
# Will place dataframes on top of each other by default df_concat=pd.concat(frames, axis=0)
# Requires a list as its main argument (this is why the frames list was created first in the example above).
# Concatenating data is also known as “setting” data in other languages like SAS. 
df_concat=pd.concat(frames)
```

```python
# Check the output
df_concat
```

```python
# Places dataframes side by side
df_concact=pd.concat(frames, axis=1)
```

```python
# Check for side by side display
df_concact
```

```python
# create a new data dictionary for trying merge command

data3={'customer_id':['6732','767','890','635'],
      'month':['Jan-20','Jan-20','Jan-20','Jan-20'],
      'days_purchased_on':[0,10,4,1]
      }

# create a dataframe df_2 from the data dictionary data3
df_2=pd.DataFrame(data3,index=[0,1,2,3])    

df_2
```

```python
# merge contents of df_2 to df dataframe based on common key customer_id
df_merged=df.merge(df_2,on='customer_id')
```

```python
df_merged
```

```python
# merge contents of df_2 to df dataframe based on common key customer_id and avoid month column appear twice
df_merged=df.merge(df_2,on=['customer_id','month'])
```

```python
df_merged
```

```python
# pandas will create an additional column in the final dataframe that indicates the source of the data within that row. 
# A value of both means the key (or keys) you specified exist in both dataframes

df_merged=df.merge(df_2,on=['customer_id','month'],indicator=True)
```

```python
df_merged
```

```python
# Test merge without overwriting
pd.merge(df,df_2, on = ['customer_id', 'month'], indicator = True)
```

```python
# the how argument specifies, as the name implies,
# how you want the dataframes to be merged, and it can take the values left, right, inner (default), or outer.

df.merge(df_2,on=['customer_id','month'],how='inner',indicator=True)
```
