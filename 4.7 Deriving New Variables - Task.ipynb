### Task 1.a. - create the “price_label” column

```python
# check the dimensions of dataframe

df_ords_prods_merge.shape
```

```python
# create a sub-set of data with 1 million records from ords_prods_merge to use in UDF

dfsubset_ords_prods_merge=df_ords_prods_merge[:1000000]
```

```python
# create a User Defined Function named 'price_label'

def price_label(row):
    
    if row['prices']<=5:
        return 'Low-range product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Mid-range product'
    elif row['prices']>15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# Add a new column 'price_range' to the dataframe sub-set

dfsubset_ords_prods_merge['price_range']=dfsubset_ords_prods_merge.apply(price_label,axis=1)
```

```python
# Look for the value counts in the column 'price_range' for each category

dfsubset_ords_prods_merge['price_range'].value_counts(dropna=False)
```

#### The for loop will run through every row in the “orders_day_of_week” column,
#### compare its value with what you know are the busiest and slowest days, and assign it the corresponding string value.
#### This information can be used to create a new column, “busiest day,”

```python
# create an empty list, 'result'
# check each value in "orders_day_of_week" column.  Based on the value append the appropriate string to the list 'result'

result=[]

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value==0:
        result.append("Busiest day")
    elif value==4:
        result.append("Least busy")
    else:
        result.append("Regular busy")
    
```

### Task 2 - Create new column “Busiest days”

```python
# print the frequency of 'orders_day_of_week' column

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
### for loop to assign the two busiest and two slowest days in the week to the 'days' list

days=[]

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value==0 or value==1:
        days.append("Most busy day")
    elif value==3 or value==4:
        days.append("Least busy day")
    else:
        days.append("Regular day")
```

```python
# check the days list

days
```

```python
# check the dataframe dimensions before adding the new column

df_ords_prods_merge.shape
```

```python
# create a new column 'Busiest days' in df_ords_prods_merge and assign the values in days list

df_ords_prods_merge['Busiest days']=days
```

### Task 4 - identify the busiest hours of the day

```python
# print the frequency of 'order_hour_of_day' column to identify the busiest hours of the day

df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False) 
```

### Based on the descriptive statistics of order_hour_of_day column
   #### values more than 75% is considered 'Most orders'
   #### values less than 25% is considered 'Fewest orders'
   #### the rest of the orders between 25% and 75% are considered 'Average orders'

```python
# describe() function gives you a quick list of descriptive statistics for a column

df_ords_prods_merge['order_hour_of_day'].describe()
```

```python
# for loop for assigning periods of time labeled “Most orders”, “Average orders" and “Fewest orders” 
# to the list 'period_of_time'


period_of_time=[]

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value==8 or value==9 or value==10 or value==11 or value==12 or value==13 or value==14 or value==15 or value==16 or value==17:
        period_of_time.append("Most orders")
    elif value==18 or value==19:
        period_of_time.append("Average orders")
    else:
        period_of_time.append("Fewest orders")
```

```python
# Look at the contents of the list

period_of_time
```

```python
# create a new column containing the labels in 'period_of_time' list

df_ords_prods_merge['busiest_period_of_day']=period_of_time
```
