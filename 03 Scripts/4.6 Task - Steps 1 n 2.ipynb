## Task 4.6- Steps 1 & 2 - Combining & Exporting Data

```python
# import libraries

import pandas as pd
import numpy as np
import os
```

### Step 1

#### Merge Instacart orders data with the new orders_products_prior

```python
# assign csv file path to a variable 'path'
path= r'C:\Users\19452\Desktop\Seema\Careerfoundry-Data Analytics Course\Work Area\Python Fundamentals\Nov-2023 Instacart Basket Analysis\02 Data'

# import orders_products_prior.csv from Original data folder and assign it to a dataframe df_orders_prior
df_ords_prior=pd.read_csv(os.path.join(path,'Original Data','orders_products_prior.csv'),index_col=False) 

# import orders_checked.csv from Prepared data folder and assign it to a dataframe df_ords
df_ords=pd.read_csv(os.path.join(path,'Prepared Data','orders_checked.csv'),index_col=False)
```

```python
# Looking at the contents of df_orders_prior dataframe

df_ords_prior
```

```python
#Looking at the dimensions of the dataframe

df_ords_prior.shape
```

```python
# Looking for the contents of df_ords dataframe

df_ords
```

```python
#Looking at the dimensions of the dataframe

df_ords.shape
```

```python
# merge df_ords_prior dataframe with df_ords dataframe using the default type of join i.e., inner
# This means that the resulting table will only contain observations found in both dataframes.
# As such, the merge flag here will only show entries that have a value of “both.”  This does not merge all the observations

df_merged_large=df_ords.merge(df_ords_prior,on='order_id',indicator=True)
```

```python
# Checking the results for merge including the merge flag

df_merged_large.head()
```

```python
# Check the dimensions of merged dataframe

df_merged_large.shape
```

```python
#Look at the value counts of the flag variable '_merge'

df_merged_large['_merge'].value_counts()
```

```python
# merge df_ords_prior dataframe with df_ords dataframe using the outer join 
# Merging like this will combine all the observations and show you the real merge rate

df_merged_large=df_ords.merge(df_ords_prior, on='order_id', how='outer',indicator=True)
```

```python
# Check the dimensions of merged dataframe

df_merged_large.shape
```

```python
# Checking the results for merge including the merge flag

df_merged_large.head()
```

```python
## Look at the value counts of the flag variable '_merge' after the outer join merge

df_merged_large['_merge'].value_counts()
```

### Step 2

#### Export the merged file in pickle format as “orders_products_combined.pkl”

```python
## Export data to pkl

df_merged_large.to_pickle(os.path.join(path,'Prepared Data','orders_products_combined.pkl'))
```
