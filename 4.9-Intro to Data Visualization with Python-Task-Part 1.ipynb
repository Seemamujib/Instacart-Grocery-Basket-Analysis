### Step 5 - Complete the fundamental data quality and consistency checks 

#### first_name seems to have mixed data types.  The current data type of first_name is object which is suitable for the
#### column.  So does not need changing anything.

```python
# See if there are any mixed data types

for col in df_customer.columns.tolist():
  weird = (df_customer[[col]].applymap(type) != df_customer[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customer[weird]) > 0:
    print (col)
```
